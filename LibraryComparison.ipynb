```python
df = pd.read_csv(foreign_path)
```

```python
df_name = df.copy()
df_name = df_name['File_Name']

df_name = df_name.to_frame()
```

```python
#Build lists of the possible seasons and episodes to look for (assuming max 99 seasons and 99 episodes in a given season)

seasons = []
episodes = []

#Build lists of the possible seasons and episodes
for i in range(0,100):
    for j in range(1,3):
        num = str(i).zfill(j)
        seasons.append(f'S{num}')
        episodes.append(f'E{num}')
seasons = list(set(seasons))
episodes = list(set(episodes))
seasons.sort()
episodes.sort()

SE = []
#Build all the permutations of season and episode combinations into a list
for se in seasons:
    for ep in episodes:
        SE.append(str(f'{se}{ep}'))
```

```python
#Build a dataframe of only tv items
tv = df_name[df_name['File_Name'].str.contains('|'.join(SE))]
movies = df_name[~df_name['File_Name'].isin(list(tv['File_Name'].values))]
```

```python
def get_showname(file_name):
    for se in SE:
        splitted = file_name.split(se)
        if splitted[0] != file_name:
            return splitted[0].strip()
```

```python
get_showname(tv.iloc[0,0])
tv['Show'] = tv['File_Name'].apply(get_showname)
```

```python
#Create a list of TV shows
tv.reset_index(drop=True, inplace=True)
movies.reset_index(drop=True, inplace=True)
shows = list(set(tv['Show'].values))
shows
```

```python
#Add episode counts to the list of TV shows
tv['Episode_Count'] = tv.groupby('Show')['Show'].transform('count')
tv
```

```python
#Build a dataframe with a uniqu list of shows and the number of episodes in each show
shows = tv[['Show','Episode_Count']]
shows.drop_duplicates()
```
